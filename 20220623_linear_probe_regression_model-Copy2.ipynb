```python
# panaliized linear regression model 
# elastic net regularization


#Dataprep
train_y_1 = target_1['swvl1'].loc['2004-10-11 16:00:00':'2009-09-16 05:00:00']
train_y_2 = target_1['swvl2'].loc['2004-10-11 16:00:00':'2009-09-16 05:00:00']
train_y_3 = target_1['swvl3'].loc['2004-10-11 16:00:00':'2009-09-16 05:00:00']
train_y_4 = target_1['swvl4'].loc['2004-10-11 16:00:00':'2009-09-16 05:00:00']

train_X_c0_schwebich = Schwebich_c0.loc[:(len(train_y_1)-1)]
train_X_c0_weierbach = Weierbach_c0.loc[:(len(train_y_1)-1)]
train_X_c0_huewelerbach = Huewelerbach_c0.loc[:(len(train_y_1)-1)]
train_X_c0_mierbech = Mierbech_c0.loc[:(len(train_y_1)-1)]
train_X_c0_bibeschbach = Bibeschbach_c0.loc[:(len(train_y_1)-1)]

test_y_1 = target_1['swvl1'].loc['2009-09-16 05:30:00':]
test_y_2 = target_1['swvl2'].loc['2009-09-16 05:30:00':]
test_y_3 = target_1['swvl3'].loc['2009-09-16 05:30:00':]
test_y_4 = target_1['swvl4'].loc['2009-09-16 05:30:00':]

test_X_c0_schwebich = Schwebich_c0.loc[len(train_y_1):]
test_X_c0_weierbach = Weierbach_c0.loc[len(train_y_1):]
test_X_c0_huewelerbach = Huewelerbach_c0.loc[len(train_y_1):]
test_X_c0_mierbech = Mierbech_c0.loc[len(train_y_1):]
test_X_c0_bibeschbach = Bibeschbach_c0.loc[len(train_y_1):]


# get date index
date_index = test_y_1.index
#date_index = date_index.set_index('Date')
#Model set up
alpha = 1.0
loss_ratio = 0.15

######### swvl 1 ########
regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
#Model run & eval
swvl1_model_c0_schwebich = regr.fit(train_X_c0_schwebich, train_y_1)
predict_c0_schwebich = regr.predict(test_X_c0_schwebich)

swvl1_model_c0_schwebich.coef_

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl1_model_c0_weierbach = regr.fit(train_X_c0_weierbach, train_y_1)
predict_c0_weierbach = regr.predict(test_X_c0_weierbach)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl1_model_c0_huewelerbach = regr.fit(train_X_c0_huewelerbach, train_y_1)
predict_c0_huewelerbach = regr.predict(test_X_c0_huewelerbach)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl1_model_c0_mierbech = regr.fit(train_X_c0_mierbech, train_y_1)
predict_c0_mierbech = regr.predict(test_X_c0_mierbech)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl1_model_c0_bibeschbach = regr.fit(train_X_c0_bibeschbach, train_y_1)
predict_c0_bibeschbach = regr.predict(test_X_c0_bibeschbach)


######### swvl 2 ########
regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
#Model run & eval
swvl2_model_c0_schwebich = regr.fit(train_X_c0_schwebich, train_y_2)
predict_c0_schwebich2 = regr.predict(test_X_c0_schwebich)

swvl2_model_c0_schwebich.coef_

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl2_model_c0_weierbach = regr.fit(train_X_c0_weierbach, train_y_2)
predict_c0_weierbach2 = regr.predict(test_X_c0_weierbach)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl2_model_c0_huewelerbach = regr.fit(train_X_c0_huewelerbach, train_y_2)
predict_c0_huewelerbach2 = regr.predict(test_X_c0_huewelerbach)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl2_model_c0_mierbech = regr.fit(train_X_c0_mierbech, train_y_2)
predict_c0_mierbech2 = regr.predict(test_X_c0_mierbech)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl2_model_c0_bibeschbach = regr.fit(train_X_c0_bibeschbach, train_y_2)
predict_c0_bibeschbach2 = regr.predict(test_X_c0_bibeschbach)

######### swvl 3 ########
regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
#Model run & eval
swvl3_model_c0_schwebich = regr.fit(train_X_c0_schwebich, train_y_1)
predict_c0_schwebich3 = regr.predict(test_X_c0_schwebich)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl3_model_c0_weierbach = regr.fit(train_X_c0_weierbach, train_y_3)
predict_c0_weierbach3 = regr.predict(test_X_c0_weierbach)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl3_model_c0_huewelerbach = regr.fit(train_X_c0_huewelerbach, train_y_3)
predict_c0_huewelerbach3 = regr.predict(test_X_c0_huewelerbach)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl3_model_c0_mierbech = regr.fit(train_X_c0_mierbech, train_y_3)
predict_c0_mierbech3 = regr.predict(test_X_c0_mierbech)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl3_model_c0_bibeschbach = regr.fit(train_X_c0_bibeschbach, train_y_3)
predict_c0_bibeschbach3 = regr.predict(test_X_c0_bibeschbach)

######### swvl 4 ########
regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
#Model run & eval
swvl4_model_c0_schwebich = regr.fit(train_X_c0_schwebich, train_y_4)
predict_c0_schwebich4 = regr.predict(test_X_c0_schwebich)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl4_model_c0_weierbach = regr.fit(train_X_c0_weierbach, train_y_4)
predict_c0_weierbach4 = regr.predict(test_X_c0_weierbach)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl4_model_c0_huewelerbach = regr.fit(train_X_c0_huewelerbach, train_y_4)
predict_c0_huewelerbach4 = regr.predict(test_X_c0_huewelerbach)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl4_model_c0_mierbech = regr.fit(train_X_c0_mierbech, train_y_4)
predict_c0_mierbech4 = regr.predict(test_X_c0_mierbech)

regr = ElasticNet(alpha = alpha, l1_ratio=loss_ratio)
swvl4_model_c0_bibeschbach = regr.fit(train_X_c0_bibeschbach, train_y_4)
predict_c0_bibeschbach4 = regr.predict(test_X_c0_bibeschbach)
```
